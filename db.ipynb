# Projeto Simples de Análise de Dados com Python (Pandas e Matplotlib).

O objetivo deste projeto é realizar uma simples análise de dados, referente a saúde física, por meio de dados brutos (.csv), podemos efetuar uma descoberta importante para um determinado público!  

 Importação da biblioteca Pandas e leitura e  do arquivo para análise (Dados Brutos)

```python
import pandas as pd
df = pd.read_csv('C:/Users/Felippe/Desktop/Gabriel/gym_members_exercise_tracking.csv')
display(df)
```

Tratamento de Dados: Renomear as colunas e linhas para o idioma português!


```python
df.rename(columns={'Age': 'Idade', 'Gender': 'Gênero', 'Weight (kg)': 'Peso (kg)', 'Height (m)': 'Altura (m)', 'Max_BPM': 'BPM_máximo','Avg_BPM': 'Media_BPM', 'Resting_BPM':'Descanso_BPM', 'Session_Duration (hours)':'Duração_Treino (horas)', 'Calories_Burned': 'Calorias_Queimadas','Workout_Type': 'Tipo_Treino','Fat_Percentage':'Porcentagem_gordura', 'Water_Intake (liters)': 'Agua_ingerida(litros)', 'Workout_Frequency (days/week)': 'Frequencia_Treino (Dias/semana)', 'Experience_Level': 'Nivel_Experiencia','BMI':'IMC'}, inplace=True)
df['Gênero'] = df['Gênero'].map({'Male': 'Masculino', 'Female': 'Feminino'})
df['Tipo_Treino'] = df['Tipo_Treino'].map({'Yoga':'Ioga','HIIT':'Alta Intensidade','Cardio':'Cardio', 'Strength':'Musculação'})
display(df)
```

Limpeza de Dados (Remover colunas desnecessárias para o contexto da análise, não tem dados duplicados ou missing values).

```python
df.info()
df.head()
df.describe()
#breve descrição de como estamos(informações sobre o tipo de dados, e descrição simples de funções em geral).
```

```python
# Índices das colunas que você deseja remover
# Substitui pelos índices corretos
colunas_para_remover_indices = [4,5,6,10,11,13]  # Exemplo de índices

# Remove as colunas usando os índices
df.drop(df.columns[colunas_para_remover_indices], axis=1, inplace=True)

# Exibe o DataFrame atualizado
display(df)
```

Média do peso por gênero.

```python
# média de peso por gênero
peso_medio= df.groupby('Gênero')['Peso (kg)'].mean()
# Exibe  DataFrame 
peso_medio_df = pd.DataFrame(peso_medio).reset_index()
peso_medio_df.columns = ['Gênero', 'Peso Médio (kg)']

# Exibe em formato tabela
print(peso_medio_df.to_string(index=False))
```

Análise sobre o grupo de homens/mulheres entre os 50 e 60 anos, com peso entre 50 a 80 Kg, e seu IMC.
 
 1 - Tabela com todas as colunas com as condições estabelecidas para ambos os gêneros.
 
 2 - Gráfico referente a situação de análise.
 
 3 - Estabelecer padrões e descobertas.

```python
# 1 - Tabela com todas as colunas com as condições estabelecidas para ambos os gêneros.
homens = df[
    (df['Gênero'] == 'Masculino') & 
    (df['Idade'] >= 50) & 
    (df['Idade'] <= 60) & 
    (df['Peso (kg)'] >= 50) & 
    (df['Peso (kg)'] <= 80) &
    (df['Frequencia_Treino (Dias/semana)'])
]
mulheres = df[
    (df['Gênero'] == 'Feminino') & 
    (df['Idade'] >= 50) & 
    (df['Idade'] <= 60) & 
    (df['Peso (kg)'] >= 50) & 
    (df['Peso (kg)'] <= 80) &
    (df['Frequencia_Treino (Dias/semana)'])
]
display(homens)
display(mulheres)
```

```python
# Calcula a média de Peso e IMC para homens
media_homens = homens[['Peso (kg)', 'IMC']].mean().round(2)

# Calcula a média de Peso e IMC para mulheres
media_mulheres = mulheres[['Peso (kg)', 'IMC']].mean().round(2)

# Exibe os resultados
print("Médias para Homens:")
print(media_homens.to_string(index=True))

print("\nMédias para Mulheres:")
print(media_mulheres.to_string(index=True))
```

```python
import matplotlib.pyplot as plt

dados = {
    'Gênero': ['Homens', 'Mulheres'],
    'Peso Médio (kg)': [media_homens['Peso (kg)'], media_mulheres['Peso (kg)']],
    'IMC Médio': [media_homens['IMC'], media_mulheres['IMC']]
}

df_medias = pd.DataFrame(dados)

# 5. Criar o gráfico de barras
fig, ax = plt.subplots(figsize=(8, 6))

# Índices para as barras
indice_barras = range(len(df_medias['Gênero']))

largura_barras = 0.2

# Gráfico de barras para Peso Médio
ax.bar([i - largura_barras/2 for i in indice_barras], df_medias['Peso Médio (kg)'], width=largura_barras, label='Peso Médio (kg)', color='royalblue')

# Gráfico de barras para IMC Médio
ax.bar([i + largura_barras/2 for i in indice_barras], df_medias['IMC Médio'], width=largura_barras, label='IMC Médio', color='lightgreen')

# Adicionando legendas e rótulos
ax.set_xlabel('Gênero')
ax.set_ylabel('Média')
ax.set_title('Média de Peso e IMC por Gênero (50-60 anos, Peso 50-80 kg)')
ax.set_xticks(indice_barras)
ax.set_xticklabels(df_medias['Gênero'])
ax.legend()

# Exibir o gráfico
plt.tight_layout()
plt.show()
```

Ao aplicar as tabelas e utilizando o artifício da visualização de dados, que facilita as apresentações deixando um estudo um pouco complexo em algo mais visual e fácil de entender, podemos observar a proximidade das médias de peso entre os grupos selecionados.

É interessante ressaltar a importância do cuidado com o corpo para esses grupos, sobretudo para o gênero feminino, que, dentro do grupo selecionado, aparece flertando com o sobrepeso!
